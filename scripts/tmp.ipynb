```python
import re
import pandas as pd


def extract_version(filename):
    match = re.search(r'(v[1-5])', str(filename).lower())
    return match.group(1).upper() if match else "V1"

df = pd.read_csv("../dataset/semantic_benchmark_with_emb.csv")
df['obfuscation'] = df['filename'].apply(extract_version)
df.to_csv("../dataset/semantic_benchmark_with_emb.csv", index=False)
```
